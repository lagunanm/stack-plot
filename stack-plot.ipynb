# StackPlot Example Using Random Numbers

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
```

Create some random dat to plot

```python
# create a range of data for x-axis
x=range(1,21)

# make two sets of random data for y-axis
y1=np.random.rand(20)
y2=np.random.rand(20)

# print values to see what has been loaded
print(x)
print(y1)
print(y2)
```

```python
# setup legend for plot
plt.plot([],[], color='r', label='y1')
plt.plot([],[], color='b', label='y2')

#setup the plot
plt.stackplot(x,y1, y2, colors=['r','b'])
plt.title('Stack Plot Random Numbers')
plt.legend()
plt.show()
```
